```python
import pandas as pd
import numpy as np
World = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
World
```

```python
Australia1 = World.loc[World["location"] == "Australia"]
Australia1
```

```python
Aus_clean1 = Australia1.fillna(0)
Aus_clean1
```

```python
import matplotlib.pyplot as plt
x = Aus_clean1['date']
y1 = Aus_clean1['total_cases']
y2 = Aus_clean1['total_deaths']
plt.plot(x, y1, label="total cases")
plt.plot(x, y2, label="total deaths")
plt.plot()
plt.xlabel("date")
plt.ylabel("Number of population")
plt.title("Line Graph Example")
plt.legend()
plt.show()
```
